```python
import sqlite3

# Подключаемся к базе данных
conn = sqlite3.connect('cars.db')  # Замените на имя вашей БД
cursor = conn.cursor()

# Выполняем запрос для подсчета автомобилей
cursor.execute("SELECT COUNT(*) FROM cars WHERE minivan == 1")  # Замените 'cars' на вашу таблицу
count = cursor.fetchone()[0]  # Получаем результат

# Выводим количество автомобилей
print(f"Всего минивэнов в базе: {count}")

# Закрываем соединение
conn.close()
```
